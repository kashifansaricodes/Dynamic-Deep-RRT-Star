```python
import os
import csv
import numpy as np
from PIL import Image

# Define the directory containing the PNG files
directory = r"C:\Users\sachi\Planning\Deep-RRT-Star-Implementation\Maps2"

# Define the output CSV file path
output_csv = "sample_points.csv"

# Define the size of the image samples
sample_size = 12

# Define the number of samples per image file
num_samples_per_image = 100

# Function to generate random sample points and orientations
def generate_sample():
    # Generate random start and goal points within the image boundaries
    start_x, start_y = np.random.randint(0, sample_size), np.random.randint(0, sample_size)
    goal_x, goal_y = np.random.randint(0, sample_size), np.random.randint(0, sample_size)
    
    # Generate random orientations (angles) in radians
    theta_s = np.random.uniform(0, 2*np.pi)
    theta_g = np.random.uniform(0, 2*np.pi)
    
    return (start_x, start_y, theta_s), (goal_x, goal_y, theta_g)

# Open or create the CSV file for writing
with open(output_csv, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["start_x", "start_y", "thetas", "goal_x", "goal_y", "thetag", "filename"])
    
    # Iterate over each PNG file in the directory
    for filename in os.listdir(directory):
        if filename.endswith(".png"):
            file_path = os.path.join(directory, filename)
            
            # Open the image file
            with Image.open(file_path) as img:
                # Convert the image to grayscale
                img_gray = img.convert("L")
                
                # Generate sample points and orientations for this image
                for _ in range(num_samples_per_image):
                    start, goal = generate_sample()
                    start_x, start_y, theta_s = start
                    goal_x, goal_y, theta_g = goal
                    
                    # Check if the start and goal points are valid (not white pixels)
                    if img_gray.getpixel((start_x, start_y)) != 255 and img_gray.getpixel((goal_x, goal_y)) != 255:
                        # Write the sample data to the CSV file
                        writer.writerow([start_x, start_y, theta_s, goal_x, goal_y, theta_g, filename])
```
