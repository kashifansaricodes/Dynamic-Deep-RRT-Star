```python
import numpy as np
import matplotlib.pyplot as plt

def create_map(width, height, obstacles=[]):
    # Create a white background
    map_array = np.ones((height, width), dtype=np.uint8) * 255
    
    # Add black obstacles
    for obstacle in obstacles:
        x, y, obstacle_width, obstacle_height = obstacle
        map_array[y:y + obstacle_height, x:x + obstacle_width] = 0

    return map_array

def save_map_to_npy(map_array, file_path):
    np.save(file_path, map_array)

def visualize_map(map_array):
    plt.imshow(map_array, cmap='gray')
    plt.show()

# Example usage:
map_width = 1200
map_height = 1200

# Initial obstacle positions and sizes
obstacles = [(400, 200, 200 , 200)]

# Create and save the map
map_array = create_map(map_width, map_height, obstacles)
#save_map_to_npy(map_array, 'Downloads/map_with_obstacles.npy')

# Visualize the map
visualize_map(map_array)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.draw import rectangle, polygon, disk

def create_map(width, height, shapes=[]):
    
    # Create a white background
    map_array = np.ones((height, width), dtype=np.uint8) * 255

    
    for shape in shapes:
        shape_type, params = shape[0], shape[1:]
        
        if shape_type == 'rectangle':
            rr, cc = rectangle(start=params[:2], end=(params[0] + params[2], params[1] + params[3]), extent=None, shape=map_array.shape)
            map_array[rr, cc] = 0
        elif shape_type == 'triangle':
            rr, cc = polygon([params[0], params[0] + params[2], params[0] + params[2]], 
                             [params[1], params[1] + params[3]//2 , params[1] - params[3]//2], shape=map_array.shape)
            map_array[rr, cc] = 0
        if shape_type == 'disk':
            cy, cx, radius = params
            rr, cc = disk((cy, cx), radius, shape=map_array.shape)
            map_array[rr , cc] = 0

    return map_array

def save_map_to_npy(map_array, file_path):
    np.save(file_path, map_array)

def visualize_map(map_array):
    plt.imshow(map_array, cmap='gray')
    plt.show()

# Example usage:
map_width = 1200
map_height = 1200

# Initial obstacle positions and sizes
# obstacles = [(400, 200, 200, 200)]

# Additional shapes (triangle, square, circle, square, circle)
shapes = [('disk', 757, 250, 146), ('rectangle', 394, 691, 228, 283), ('rectangle', 405, 335, 243, 251), ('triangle', 610, 209, 266, 228), ('disk', 440, 179, 122)]


# Create and save the map
map_array = create_map(map_width, map_height, shapes)
#save_map_to_npy(map_array, 'Downloads/map_with_shapes.npy')

# Visualize the map
visualize_map(map_array)
```

```python
import csv
import random

# Function to generate random values within given constraints for each shape
def generate_random_values(shape):
    if shape == 'triangle':
        tx = random.randint(194, 1006)
        ty = random.randint(5, 857)
        th = random.randint(173, 338)
        tb = random.randint(200, 390)
        return tx, ty, th, tb
    elif shape == 'rectangle':
        rx = random.randint(5, 879)
        ry = random.randint(321, 879)
        rh = random.randint(187, 316)
        rb = random.randint(187, 316)
        return rx, ry, rh, rb
    elif shape == 'disk':
        cx = random.randint(165, 1035)
        cy = random.randint(165, 1035)
        rad = random.randint(105, 160)
        return cx, cy, rad

# List of shapes
shapes = ['triangle', 'rectangle', 'disk']

filepath = r"C:\Users\sachi\Planning\Deep_RRT\shapes.csv"
# Generate data and write to CSV file
with open(filepath, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Shape', 'Parameter 1', 'Parameter 2', 'Parameter 3', 'Parameter 4'])

    # Generate 500 rows of data
    for _ in range(500):
        shape = random.choice(shapes)
        params = generate_random_values(shape)
        writer.writerow([shape] + list(params))
```

```python
import numpy as np
from skimage.draw import rectangle, polygon, disk
import csv
import os

def create_map(width, height, shapes=[]):
    # Create a white background
    map_array = np.ones((height, width), dtype=np.uint8) * 255
    
    for shape in shapes:
        shape_type, params = shape[0], shape[1:]
        
        if shape_type == 'rectangle':
            rr, cc = rectangle(start=params[:2], end=(params[0] + params[2], params[1] + params[3]), extent=None, shape=map_array.shape)
            map_array[rr, cc] = 0
        elif shape_type == 'triangle':
            rr, cc = polygon([params[0], params[0] + params[2], params[0] + params[2]], 
                             [params[1], params[1] + params[3]//2 , params[1] - params[3]//2], shape=map_array.shape)
            map_array[rr, cc] = 0
        elif shape_type == 'disk':
            cy, cx, radius = params
            rr, cc = disk((cy, cx), radius, shape=map_array.shape)
            map_array[rr , cc] = 0

    return map_array

def write_maps_to_npy(csv_file, output_folder, width, height):
    with open(csv_file, mode='r') as file:
        csv_reader = csv.reader(file, delimiter='\t')
        next(csv_reader)  # Skip header row
        map_count = 1
        shapes = []
        for row in csv_reader:
            shape_type = row[0]
            params = tuple(map(int, row[1:]))  # Convert parameters to integers
            shapes.append((shape_type,) + params)
            if len(shapes) == 5:
                # Generate map and write to file
                map_array = create_map(width, height, shapes)
                np.save(os.path.join(output_folder, f'map_{map_count}.npy'), map_array)
                shapes = []  # Reset shapes for the next set of rows
                map_count += 1

# Define input CSV file and output folder
csv_file = r"C:\Users\sachi\Planning\Deep_RRT\shapes.csv"
output_folder = r"C:\Users\sachi\Planning\Deep_RRT\Maps"

# Define map dimensions
width = 1200
height = 1200

# Create output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Write maps to .npy files
write_maps_to_npy(csv_file, output_folder, width, height)
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt

def visualize_npy_files(folder_path, num_files):
    npy_files = [file for file in os.listdir(folder_path) if file.endswith('.npy')]
    npy_files.sort()  # Sort the files to maintain order
    
    num_files = min(num_files, len(npy_files))  # Ensure we don't exceed the available number of files
    
    for i in range(num_files):
        file_path = os.path.join(folder_path, npy_files[i])
        map_array = np.load(file_path)
        plt.figure(figsize=(6, 6))
        plt.imshow(map_array, cmap='gray')
        plt.title(f"Map from {npy_files[i]}")
        plt.show()

# Example usage:
folder_path = r"C:\Users\sachi\Planning\Deep_RRT\Maps"
num_files_to_visualize = 5  # Change this number to visualize a different number of files
visualize_npy_files(folder_path, num_files_to_visualize)
```

```python
import os
import numpy as np
import csv

def generate_start_goal_points(npy_folder, output_csv, num_points):
    npy_files = [file for file in os.listdir(npy_folder) if file.endswith('.npy')]
    npy_files.sort()  # Sort the files to maintain order
    
    with open(output_csv, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['File', 'Start X', 'Start Y', 'Goal X', 'Goal Y'])
        
        for npy_file in npy_files:
            file_path = os.path.join(npy_folder, npy_file)
            map_array = np.load(file_path)
            
            height, width = map_array.shape
            obstacles = np.where(map_array == 0)  # Get the indices of black pixels
            
            for _ in range(num_points):
                start = None
                goal = None
                
                while start is None or goal is None or start == goal:
                    start = (np.random.randint(width), np.random.randint(height))
                    goal = (np.random.randint(width), np.random.randint(height))
                    
                    if start in obstacles[0] or goal in obstacles[0]:
                        start = None
                        goal = None
                
                writer.writerow([npy_file, start[0], start[1], goal[0], goal[1]])

# Example usage:
npy_folder = r"C:\Users\sachi\Planning\Deep-RRT-Star-Implementation\Maps"
output_csv = r" C:\Users\sachi\Planning\Deep-RRT-Star-Implementation\start_goal_points.csv"
num_points = 1000

generate_start_goal_points(npy_folder, output_csv, num_points)
```
